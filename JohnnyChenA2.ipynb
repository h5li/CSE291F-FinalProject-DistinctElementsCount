```python
import numpy as np
import math
import random
from JohnnyChen import *
import gmpy
import pandas as pd
from tqdm import tqdm
# Load data
airbnb = np.load('./data/airbnb_location.npy')
google = np.load('./data/google_app.npy')
movie_id = np.array(pd.read_csv('./data/ml_small_movieids.csv', header=None)[0])
user_id = np.array(pd.read_csv('./data/ml_small_userids.csv', header=None)[0])
def test_A2(elements,m):
    np.random.shuffle(elements)
    success = 0
    result_list = []
    true_r = len(set(elements))
    for i in tqdm(range(100)):
        r = A2(elements,m,0.5)
        result_list.append(r)
        if r >= 0.5*true_r and r <= 1.5*true_r:
            success += 1
    return true_r, np.mean(result_list),np.var(result_list),success
# Test Airbnb
def test_A2_median_of_means(elements,m):
    np.random.shuffle(elements)
    success = 0
    result_list = []
    true_r = len(set(elements))
    for i in tqdm(range(100)):
        sub_r = []
        for i in range(5):
            sum_r = 0
            for j in range(20):
                r = A2(elements,m,0.5)
                sum_r += r
            sub_r.append(sum_r/20)
        result_list.append(np.median(sub_r))
        if np.median(sub_r) >= 0.5*true_r and np.median(sub_r) <= 1.5*true_r:
            success += 1
    return true_r, np.mean(result_list),np.var(result_list),success
```

```python
print(test_A2(airbnb,10000))
print(test_A2_median_of_means(airbnb, 10000))
print(test_A2(google,1500))
print(test_A2_median_of_means(google,1500))
print(test_A2(movie_id,10000))
print(test_A2_median_of_means(movie_id,10000))
print(test_A2(user_id,1000))
print(test_A2_median_of_means(user_id,1000))
```
